```python
import pandas as pd
```

```python
# 1. Create the initial list
raw_data = [1, 3, 9, 10, 12, 15, 18, 20]
```

```python
# 2. Create DataFrame and filter for even values
df = pd.DataFrame(raw_data, columns=['value'])
df_even = df[df['value'] % 2 == 0].copy()
```

```python
# 3. Calculate the sum using Python/Pandas
even_sum = df_even['value'].sum()
```

```python
print(f"Even Numbers Found: {df_even['value'].tolist()}")
print(f"Mathematical Sum: {even_sum}")
```

```python
import vertexai
from google.genai import Client, types

# 1. Initialize the Vertex AI environment
PROJECT_ID = "agntworks-dev"  # Use your actual project ID
LOCATION = "us-central1"      # or your preferred region

# 2. Create the Client
# In Vertex AI Workbench, this usually handles authentication automatically
client = Client(vertexai=True, project=PROJECT_ID, location=LOCATION)

print(f"Gemini client initialized for project: {PROJECT_ID}")
```

```python
even_list = df_even['value'].tolist()
```

```python
user_prompt = f"Sum these numbers and provide the result only: {even_list}"
```

```python
response = client.models.generate_content(
    model="gemini-2.5-pro",
    contents=[
        types.Content(role="user", parts=[types.Part(text=user_prompt)])
    ]
)

print("--- Gemini's Calculation ---")
print(response.text.strip())
```
