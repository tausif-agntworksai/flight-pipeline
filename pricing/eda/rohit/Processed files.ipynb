```python
# 1. Get a simple array of all unique cluster names
unique_clusters = df3['cluster'].unique()

# 2. Get a sorted distribution (how many airports per cluster)
# This helps you see the "Operational Footprint" of each cluster
cluster_distribution = df3['cluster'].value_counts().reset_index()
cluster_distribution.columns = ['Cluster Name', 'Airport Count']

# 3. Print the results
print(f"Total Unique Clusters Found: {len(unique_clusters)}")
print("-" * 30)
print(cluster_distribution)

# Optional: If you want to see them alphabetically
# print(sorted(unique_clusters))
```
