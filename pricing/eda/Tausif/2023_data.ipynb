```python
import pandas as pd
import fsspec
from zipfile import ZipFile
from io import BytesIO
def load_specific_csv_from_zip(url, filename):
    """Load one specific CSV from multi-file ZIP"""
    with fsspec.open(url, 'rb') as f:
        with ZipFile(BytesIO(f.read())) as z:
            return pd.read_csv(z.open(filename))

url = "gs://agntworks-data-dev/wheelsup/raw/AgntWorks.zip"
data = load_specific_csv_from_zip(url,'AgntWorks/booked_flights.csv')
```

```python
df = data.copy()
df.columns
```

```python
df['time'] = pd.to_datetime(df['flightActualDepartureTime'],utc=True)
df['hour'] = df['time'].dt.hour
```

```python
# 1. Define the cabin categories
mid_cabins = ['Super Midsize', 'Premium Super-Mid']
light_cabins = ['Light', 'Premium Light']
target_cabins = mid_cabins + light_cabins

# 2. Apply filters: flightStatus must be 'Flown' AND cabin must be in our target list
df_flown = df[
    (df['flightStatus'] == 'Flown') & 
    (df['flightactualAircraftCabinName'].isin(target_cabins))
].copy()

# 3. Summary of results
print(f"✅ df_flown created successfully.")
print(f"Rows after filtering: {len(df_flown)} (from {len(df)} original rows)")
print("\nCabin counts in new df:")
print(df_flown['flightactualAircraftCabinName'].value_counts())
```

```python
# 1. Define your filtered subset first

numerator = df_flown[df_flown['hour'].between(0, 6)]['flightId'].count()
denominator = df_flown['flightId'].count()

percentage = (numerator / denominator) * 100
print(f"Percentage of Night/Early Morning Flights: {percentage:.2f}%")
```
