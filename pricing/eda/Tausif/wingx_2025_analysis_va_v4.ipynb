```python
import pandas as pd
import fsspec
from zipfile import ZipFile
from io import BytesIO

def load_specific_csv_from_zip(url, filename):
    """Load one specific CSV from multi-file ZIP"""
    with fsspec.open(url, 'rb') as f:
        with ZipFile(BytesIO(f.read())) as z:
            return pd.read_csv(z.open(filename))
yr='25'
url = f"gs://agntworks-data-dev/wheelsup/raw/wingx/WingX_20{yr}.zip"

# Create quarter-wise DataFrames (exactly matching your file names)
df_2025_Q1 = load_specific_csv_from_zip(url, f'WINGX_Jan{yr}-Mar{yr}.csv')
df_2025_Q2 = load_specific_csv_from_zip(url, f'WINGX_Apr{yr}-Jun{yr}.csv')
df_2025_Q3 = load_specific_csv_from_zip(url, f'WINGX_Jul{yr}-Sep{yr}.csv')
df_2025_Q4 = load_specific_csv_from_zip(url, f'WINGX_Oct{yr}-Dec{yr}.csv')

# Combine ALL quarters into master df
df = pd.concat([df_2025_Q1, df_2025_Q2, df_2025_Q3, df_2025_Q4], ignore_index=True)

# Verify
print("Q1 shape:", df_2025_Q1.shape)
print("Q2 shape:", df_2025_Q2.shape) 
print("Q3 shape:", df_2025_Q3.shape)
print("Q4 shape:", df_2025_Q4.shape)
print("FULL df shape:", df.shape)
print("\nFirst few rows of combined df:")
df.head()
```

```python
# 1. Define the core parameters
target_types = ['Branded Charter', 'Aircraft Management']
strategic_segments = ['Light Jet', 'Super Light Jet', 'Super Midsize Jet']

# 2. Filter for Revenue-only flights within the core operator types and segments
# We use 'N' to ensure we are only looking at flights with passengers/revenue
revenue_core_df = df[
    (df['reposition_flight'] == 'N') & 
    (df['operator_type'].isin(target_types)) &
    (df['aircraft_segment'].isin(strategic_segments))
].copy()

# 3. Group by Operator Type to compare Wheels Up vs the Total Market
# 'is_wheels_up' is the boolean flag we created earlier for their entities
share_analysis = revenue_core_df.groupby('operator_type').agg(
    Total_Market_Revenue_Flights=('operator_type', 'count'),
    WU_Revenue_Flights=('is_wheels_up', 'sum')
)

# 4. Calculate the real Market Share Percentage
share_analysis['Market_Share_Percent'] = (
    share_analysis['WU_Revenue_Flights'] / share_analysis['Total_Market_Revenue_Flights']
) * 100

print("--- Wheels Up Revenue Market Share (Core Segments Only) ---")
print("Segments: Light, Super Light, and Super Midsize")
print("-" * 55)
print(share_analysis.sort_values(by='Market_Share_Percent', ascending=False))

# 5. Calculate overall revenue share across these two buckets
total_revenue_flights = share_analysis['Total_Market_Revenue_Flights'].sum()
total_wu_revenue_flights = share_analysis['WU_Revenue_Flights'].sum()
overall_share = (total_wu_revenue_flights / total_revenue_flights) * 100

print("-" * 55)
print(f"Overall WU Revenue Share in these Categories: {overall_share:.2f}%")
```

```python
import pandas as pd

# ── 0. Select only needed columns & cast categoricals to str ────────────────
cols = ['FlightDate_utc', 'Operator', 'Hours', 'operator_type',
        'aircraft_type', 'aircraft_segment', 'from_cluster', 'to_cluster']
df1 = df[cols].copy()
```

```python
df1 = df1[df1['aircraft_segment']=='Light Jet']
```

```python
# Cast ALL category columns to str in one go
cat_cols = ['Operator', 'operator_type', 'aircraft_type',
            'aircraft_segment', 'from_cluster', 'to_cluster']
df1.loc[:,cat_cols] = df1.loc[:,cat_cols].astype(str)
```

```python
# # ── 1. Filter operators of interest ─────────────────────────────────────────
# valid_operators = ["Wheels Up Private Jets", "Wheels Up LLC",
#                    "Mountain Aviation", "Alante Air Charter"]
# df1 = df1[df1['Operator'].isin(valid_operators)].copy()
```

```python
# ── 2. Date fields ───────────────────────────────────────────────────────────
df1['FlightDate_utc'] = pd.to_datetime(df1['FlightDate_utc'], utc=True)
df1['month_period']   = df1['FlightDate_utc'].dt.to_period('M')
df1['month_str']      = df1['month_period'].astype(str)
df1['month']          = df1['FlightDate_utc'].dt.month
df1['year']           = df1['FlightDate_utc'].dt.year
```

```python
# ── 3. Normalize route pair (alphabetical so A↔B == B↔A) ────────────────────
df1['cluster_a']  = df1.apply(lambda r: min(r['from_cluster'], r['to_cluster']), axis=1)
df1['cluster_b']  = df1.apply(lambda r: max(r['from_cluster'], r['to_cluster']), axis=1)
df1['route_pair'] = df1['cluster_a'] + ' ↔ ' + df1['cluster_b']
```

```python
# True = A→B (outbound), False = B→A (inbound)
df1['is_outbound'] = df1['from_cluster'] == df1['cluster_a']

# Drop same-cluster hops (repositions like ATLANTA ↔ ATLANTA)
df1 = df1[df1['cluster_a'] != df1['cluster_b']].copy()

# Hours to float64 (was float32)
df1['Hours'] = df1['Hours'].astype('float64')

print(f"Working dataset: {len(df1):,} rows | "
      f"{df1['route_pair'].nunique()} unique route pairs")
```

```python
# ── 4. Monthly imbalance ─────────────────────────────────────────────────────
monthly_grp = (
    df1.groupby(
        ['route_pair', 'cluster_a', 'cluster_b', 'month_period', 'month_str', 'is_outbound'],
        observed=True
    )['Hours']
    .agg(total_hrs='sum', flight_count='count', avg_hrs='mean')
    .reset_index()
)

monthly_pivot = monthly_grp.pivot_table(
    index=['route_pair', 'month_period', 'month_str'],
    columns='is_outbound',
    values=['total_hrs', 'flight_count', 'avg_hrs']
).reset_index()
```

```python
# Flatten columns
# Step 1: Flatten MultiIndex columns safely
monthly_pivot.columns = [
    '_'.join([str(c) for c in col if c != '']) if isinstance(col, tuple) else col
    for col in monthly_pivot.columns
]

# Step 2: Rename columns properly (based on actual pivot output)
monthly_pivot = monthly_pivot.rename(columns={
    'avg_hrs_False': 'inbound_avg_hrs',
    'avg_hrs_True': 'outbound_avg_hrs',
    'flight_count_False': 'inbound_count',
    'flight_count_True': 'outbound_count',
    'total_hrs_False': 'inbound_total_hrs',
    'total_hrs_True': 'outbound_total_hrs'
})

# Step 3: Extract cluster_a and cluster_b from route_pair
monthly_pivot[['cluster_a', 'cluster_b']] = (
    monthly_pivot['route_pair'].str.split(' ↔ ', expand=True)
)

# Step 4: Fill missing values
fill_cols = [
    'inbound_total_hrs', 'outbound_total_hrs',
    'inbound_count', 'outbound_count',
    'inbound_avg_hrs', 'outbound_avg_hrs'
]

monthly_pivot[fill_cols] = monthly_pivot[fill_cols].fillna(0)

# Step 5: Calculate imbalance
monthly_pivot['imbalance_hrs'] = (
    monthly_pivot['outbound_total_hrs'] - monthly_pivot['inbound_total_hrs']
)

monthly_pivot['abs_imbalance_hrs'] = monthly_pivot['imbalance_hrs'].abs()

# Step 6: Direction interpretation
monthly_pivot['imbalance_direction'] = monthly_pivot.apply(
    lambda r: (
        f"{r['cluster_a']} → {r['cluster_b']} heavier"
        if r['imbalance_hrs'] > 0.05 else
        f"{r['cluster_b']} → {r['cluster_a']} heavier"
        if r['imbalance_hrs'] < -0.05 else
        'Balanced'
    ),
    axis=1
)
```
